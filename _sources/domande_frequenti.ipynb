- **Come possono essere utilizzate le competenze acquisite nel Corso nella tua attività professionale?**

    - *Corrado Caudek.* Nel caso delle professioni d’aiuto, una delle questioni cruciali riguarda la perdita di senso che può derivare da una crisi, sia essa personale, economica, culturale, sociale, o di qualsiasi altra natura. Propongo che la risposta a questa perdita di senso sia di tipo spirituale, non legata a una Chiesa o a dogmi come la Scienza, il Piacere o la Politica, ma a una spiritualità che emerge, ad esempio, osservando le opere di Anselm Kiefer. Le opere di Kiefer evocano l’Assenza del Sacro, non come qualcosa che esisteva e abbiamo dimenticato, ma come qualcosa che la Scienza ha dimostrato essere un’illusione infantile, un’Assenza di qualcosa che non è mai esistito. Questa Assenza è fondamentale e può rispondere alla perdita di senso. La riflessione Madhyamaka del Buddhismo tibetano, proposta 2000 anni fa, suggerisce che questa Assenza è il Nirvana. Meditare su questo paradosso è la risposta più vicina che conosco alla perdita di senso. La ricaduta professionale del mio intervento consiste nel tradurre la riflessione Madhyamaka in un linguaggio contemporaneo, vicino alla psicologia. In sintesi, il mio intervento mira a fornire agli operatori delle professioni d’aiuto strumenti per comprendere e affrontare la perdita di senso attraverso una prospettiva spirituale che trascende le convenzioni religiose e abbraccia una visione più profonda e universale dell’esistenza.

- **Come si raggiunge Pomaia?**

*Treno* La stazione più vicina a Pomaia, dove si trova l'Istituto, è Rosignano, in cui si arriva con i treni regionali. I treni più veloci fermano a Cecina, che dista 20 km circa. 

Su richiesta all'Istituto è possibile prenotare una navetta da Rosignano a Pomaia (e viceversa).

*Auto* Come raggiungere Pomaia in auto:

- Per chi arriva da sud: imboccare la Strada Statale SS206 direzione Pisa. Passata la località Badie, all’altezza di una sopraelevata, girare a destra in prossimità dell’indicazione per Pomaia o per Knauf. Proseguire per 7 km. Poche centinaia di metri prima di arrivare in paese a Pomaia, girare a destra in prossimità di una strada sterrata caratterizzata da un arco tibetano all’imbocco.

- Per chi arrva da nord: imboccare la Strada Statale SS206 direzione Cecina-Roma, a circa 12 km da Cecina, all’altezza di una sopraelevata, girare a sinistra in prossimità dell’indicazione per Pomaia o per Knauf. Proseguire per 7 km. Poche centinaia di metri prima di arrivare in paese a Pomaia, girare a destra in prossimità di una strada sterrata caratterizzata da un arco tibetano all’imbocco.

*Negli anni passati, sono stati spesso organizzati gruppi per condividere il viaggio in auto.*
